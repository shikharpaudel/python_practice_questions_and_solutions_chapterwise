## 4. (Arithmetic) for each of the arithmetic operators +,-,*,/.// and ** , display the value of an expression with 27.5 as the left operand and 2 as the right operand.

```python
# first assign value to the left and right operand
left_op = 2
right_op = 27.5
# displaying value for + sign ..we will assign total variable to store sum ..you can also print directly like print(left_op + right_op)
total = left_op + right_op
print(total)
# displaying value for - sign ..we will assign remain variable to store substraction ..you can also print directly like print(left_op - right_op)
remain = left_op - right_op
print(remain)
# displaying value for * sign ..we will assign multiple variable to store multiplication ..you can also print directly like print(left_op * right_op)
multiple = left_op * right_op
print(multiple)
# displaying value for / sign ..we will assign div variable to store division  ..you can also print directly like print(left_op / right_op) NOTE: / this returns you a floating output
div = left_op / right_op
print(div)
# displaying value for // sign ..we will assign intdiv variable to store division  ..you can also print directly like print(left_op // right_op) NOTE: // this returns you a integer output discarding the fractional value.
div = left_op // right_op
print(div)
# displaying value for ** sign ..we will assign power variable to store power  ..you can also print directly like print(left_op ** right_op).
power = left_op ** right_op
print(power)
```

## 12. (7% Investment Return) Some investment advisors say that it's reasonable to expect 7% return over a long term in the stock market. Assuming that you begin with $1000 and leave your money instead, calculate and display how much money you'll have after 10,20,30 years. Use the following formulae for determining these amounts:
## a = p(1+r)^n
## where p is the original amount invested
## r is the annual rate of return
## n is the number of years
## a is the amount in deposite at the end of nth year

```python
# Given values
p = 1000        # initial investment
r = 0.07        # 7% annual return
#ROI calculations
amount_10 = p * (1 + r) ** 10
amount_20 = p * (1 + r) ** 20
amount_30 = p * (1 + r) ** 30

#results
print(f"Amount after 10 years: ${amount_10:.2f}")
print(f"Amount after 20 years: ${amount_20:.2f}")
print(f"Amount after 30 years: ${amount_30:.2f}")
```

## 13. (Target Heart-Rate Calculator) While exercising, you can use a heart-rate monitor to see that your heart rate stays within safe range suggested by your doctors and trainers. According to the American Heart Association (AHA), The formulae for calculating your maximum heart rate in beats per minute is 220 minus your age in years. your target heart rate is 50-85% of your maximum heart rate. write a script that prompts for and input the user's age and calculates and displays the users maximum heart rate and the range of the user's target heart rate.

```python
age = int(input("please enter your age:"))
#calculating maximum heart rate per minute
maximum_heart_rate_per_minute = 220-age
# calculating range of the users target heart rate
low_range = (maximum_heart_rate_per_minute * 50)/ 100
high_range = (maximum_heart_rate_per_minute * 85)/ 100
#display result
print(f"Your Maximum Heart rate is {maximum_heart_rate_per_minute}")
print(f"you target heart rate range is {low_range} - {high_range}")
```
